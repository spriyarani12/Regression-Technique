```python
import pandas as pd
import matplotlib.pyplot as plt
# use for regression visiualization
import seaborn as sns
```

```python
df=pd.read_csv('temperatures.csv')
```

```python
df
```

```python
df.head()
```

```python
# input data
x=df['YEAR']

#output data
y=df['ANNUAL']
```

```python
plt.figure(figsize=(16,9))
plt.title('temperatures plot of INDIA')
plt.xlabel('Year')
plt.ylabel('Annual average temperatures')
plt.scatter(x,y)
```

```python
x.shape
```

```python
x=x.values
```

```python
x
```

```python
x=x.reshape(117,1)
```

```python
x
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor=LinearRegression()
```

```python
regressor.fit(x,y)
```

```python
regressor.coef_
```

```python
regressor.intercept_
```

```python
predicted=regressor.predict(x)
```

```python
predicted
```

```python
import numpy as np
```

```python
np.mean(abs(y - predicted))
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y,predicted)
```

```python

np.mean(abs(y - predicted)**2)
```

```python
# find r2 square metrics
from sklearn.metrics import r2_score
r2_score(y,predicted)
```

```python
#plt.figure(figsize=(16,9))
plt.title('temperatures plot of INDIA')
plt.xlabel('Year')
plt.ylabel('Annual average temperatures')
plt.scatter(x,y,label='actual',color='purple')
plt.plot(x,predicted,label='predicted',color='g')
plt.legend()
```
